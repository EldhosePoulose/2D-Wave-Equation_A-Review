Modify previous example and define function for testing if the number is a prime number.

```python
def testPrime(num):
  # define a flag variable
  flag = False
  #prime numbers are greater than 1
  if num > 1:
      for i in range(2, num):
          if (num % i) == 0:
              flag = True
              break

  # check if flag is True
  if flag:
      return print(num, "is not a prime number")
  else:
      return print(num, "is a prime number")
```

# Task 8

The sigmoid function is frequently used activation of a neuron. It is defined by

$$y=\sigma(x)=\frac{1}{1+e^{-x}}$$ 

Draw a graf of this function for $x\in <-8,8>$. 

```python

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def relu(x):
    return(np.maximum(0, x))

def gaussian(x, sig):
    return np.exp(-np.power(- x, 2.) / (2 * np.power(sig, 2.)))

x = np.arange(-8., 8., 0.2)  
p = sigmoid(x)

plt.subplots(figsize=(12, 6))
plt.plot(x, p, label="$𝜎(𝑥)=1/1+𝑒^𝑥$") 
plt.xlabel("x") 
plt.ylabel("y")  
plt.title("The sigmoid function")
plt.legend(loc='center right')
#plt.grid(True, linestyle='--', alpha=2)
plt.axhline(y = 0, color = 'grey', linestyle="--")
plt.axhline(y = 0.5, color = 'grey', linestyle="--")
plt.axhline(y = 1, color = 'grey', linestyle="--")
plt.axvline(x = 0, color = 'black')
plt.show()
```

# Task 12

Draw response of the two input perceptron neuron defined by

$$y=\sigma(\sum_{i=0}^{1}w_ix_i + b)$$

where $w_0=10$,$w_1=10$, $b=-10$.

```python
# function
# Plot 
# meshgrid
# output
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
w_0 = 10
w_1 = 10
x_0 = np.linspace(0, 1, 100)
x_1 = np.linspace(0, 1, 100)
b = -10
def two_in_perceptron(x, y):
    return sigmoid((w_0 * x + w_1*y) +b)

X, Y = np.meshgrid(x_0, x_1)
Z = two_in_perceptron(X, Y)

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="magma")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()

```
